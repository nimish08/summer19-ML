```
import tweepy
```

```
consumerkey=''
consumersecret=''
accessToken=''
accessTokenSecret=''

auth=tweepy.OAuthHandler(consumer_key=consumerkey,consumer_secret=consumersecret)
auth.set_access_token(accessToken,accessTokenSecret)
```

```
api=tweepy.API(auth)
```

```
search=input("Enter word to be searched")
numberwords=int(input("Number of words"))
```

```
result=api.search(search,count=numberwords)
result
```

```
for i in result:
  print(i.text)
```
